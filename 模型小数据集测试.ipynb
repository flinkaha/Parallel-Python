```python
import pickle
import pandas as pd
import numpy as np

with open("w2c_sg_hs_live_sum_message.pickle","rb") as f:
    model = pickle.load(f)

data = pd.read_pickle("live_message_sum_1w.pickle")
data = data[1:3]

def predict_proba(oword, iword):
    iword_vec = model[iword]
    oword = model.wv.vocab[oword]
    oword_l = model.syn1[oword.point].T
    dot = np.dot(iword_vec, oword_l)
    lprob = -sum(np.logaddexp(0, -dot) + oword.code * dot)
    return lprob

from collections import Counter
def keywords(s):
    s = [w for w in s if w in model]
    ws = {w: sum([predict_proba(u, w) for u in s]) for w in s}
    a=Counter(ws).most_common()
    a=a[:20]
    return [i[0] for i in a]

import time
start = time.time()
import jieba
print("begin to extract keywords......")
data['live_content_sum'] = data['live_content_sum'].map(lambda x: keywords(jieba.cut(str(x[:10000]))))


with open('live_message_sum_1w_keywords.pickle', 'wb') as f:
    pickle.dump(data,f)
print(".....done.....")
end = time.time()
print("cost time:"+str(end-start))
```

```python
def isprime(n):
    if not isinstance(n,int):
        raise TypeError("n wrong type ")
    if n <2:
        return False
    if n==2:
        return True
    max=int(math.ceil(math.sqrt(n)))
    i=2
    while i <=max:
        if n % i ==0:
            return False
        i+=1
    return True

def sum_primes(n):
    return sum([x for x in range(2,n) if isprime(x)])

ppserver=()
job_server = pp.Server(ppservers=())
print ("Starting pp with", job_server.get_ncpus(), "workers")
###Starting pp with 8 workers
job1 = job_server.submit(sum_primes, (100,), (isprime,), ("math",))
result = job1()
print ("Sum of primes below 100 is", result)
####Sum of primes below 100 is 1060
```

```python
start=time.time()
inputs=(100000, 100100, 100200, 100300, 100400, 100500, 100600, 100700)
jobs = [(input, job_server.submit(sum_primes,(input,), (isprime,), ("math",))) for input in inputs]
for input, job in jobs:
    print ("Sum of primes below", input, "is", job())

print ("Time elapsed: ", time.time() - start, "s")
job_server.print_stats()

```
